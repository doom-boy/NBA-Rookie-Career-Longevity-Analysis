# **Setup / Initial**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
#pd.set_option('display.width', 500)

df = pd.read_csv('dataset/nba_logreg.csv')
df.head()
```

```python
df.info()
```

column '3p' has 1329 non-null. That is 11 less than normal. Let's see why that is.

```python
null_3p = df[df['3p'].isnull()]
display(null_3p)
```

Null values in the '3p' (3-point percentage) column means the player has **No 3-pt attempts:** ('3pa' = 0). This is true for all 11 instance in this case.

Will change it to be 0 instead of null.

```python
df['3p'] = df['3p'].fillna(0)
df.info()
```

```python
df.duplicated().sum()
```

```python
duplicate_rows = df[df.duplicated(keep=False)]

if not duplicate_rows.empty:
    display(duplicate_rows.sort_values(by=list(df.columns)))
```

Need to drop duplicated rows. Some names like Charles Smith have 3 different stat variations. Looking into this, this information is correct. There are 3 different Charles Smith that appeared at different times. Same goes with Charles Jones.

```python
print(f"Number of rows before: {len(df)}")
```

```python
df.drop_duplicates(inplace=True)
print(f"Number of rows after dropping duplicates: {len(df)}")
df.info()
```

```python
df.describe()
```

Output new, cleaned CSV file

```python
import os

output_dir = 'dataset'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

output_filepath = os.path.join(output_dir, 'nba_cleaned.csv')
df.to_csv(output_filepath, index=False)
print(f"Cleaned data saved to {output_filepath}")
```
